```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Universities.csv')
plt.hist(df['SAT'])
plt.xlabel('SATscore')
plt.ylabel('SATscore')
plt.title('Histogram of SATScore')
plt.show()
```

```python
plt.hist(df['Accept'])
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(df['SAT'])
```

```python
plt.plot(df['SAT'],df['Accept'],'rs')
plt.xlabel('SAT')
plt.ylabel('Accept')
plt.title('Scatterplt SAT vs Accept')
plt.show()
```

```python
plt.plot(df['SAT'],df['Expenses'],'rs')
```
